<font size=5>Sorting the Data</font>

```python
# Sorting the data by the values along either axis.
# Getting first 10 rows sorted according to the 'Fare' in non-decreasing order
df.sort_values('Fare').head(10)
```

```python
# Getting first 10 passengers who paid the highest ticket fares by using sort_values method with ascending = False attribute
# Here original data is not sorted
df.sort_values('Fare', ascending = False).head(10)
```

```python
# Above sorting did not sorted the original dataset.
# Sorting DataFrame rows in descending order on "Fare" and using  inplace = True attribute
df.sort_values("Fare", ascending = False, inplace = True)
df.head()
```

```python
# Sorting Cabin column which has Missing (Nan) Values.
# By using na_postion = 'last' puts Nan at the end. default is 'last'
df.sort_values('Cabin', ascending = False, na_position = 'last').tail()
```

<font size=5>Getting unique values in a DataFrame</font>

```python
# Finding the unique values that occur in dataset or in a column
# Using .nunique() on dateset
df.nunique()
```

```python
# Using .nunique() on particular column 'Embarked'
df["Embarked"].nunique() 
```

<font size=5>After checking data types of column in data set by using 'df.dtypes', it is observed the Embarked column is object.
After counting the unique values in Embarked column with .unique(), we can see that there are 3 unique values in that column.
Hence it can be consider that its data type should be Categorical.</font>

```python
# Changing Embarked column data tpe to Categorical by using .astype("category")
df['Embarked'] = df['Embarked'].astype('category')
df['Embarked'].dtype
```

```python
# Filtering - finding passengers who Embarked from port 'Q'
df[df["Embarked"] == "Q"].head()
```

```python
# Filtering - Two or more conditions using AND, OR opertor
# Finding Female passengers having ticket fare < 100
df_fare_mask = df["Fare"] < 100
df_gender_mask = df["Sex"] == "female"
df[df_fare_mask & df_gender_mask].head()
```

<font size=5> Checking if any value is Nan in a Dataframe</font>

```python
# Finding Null Values using .isnull()
# Finding passengers whose is not known i.e Nan
null_mask = df["Cabin"].isnull() 
df[null_mask].head()
```

```python
# Finding the count of the passengers who cabin is not known using .isnull.sum() functions
df['Cabin'].isnull().sum()
```

<font size=5> As there are lot of *missing values* in the dataset columns. Hence, dropping 'Cabin' column which has 70% missing values.</font>


```python
# Droping the 'Cabin' column 
df_without_cabin = df.drop(labels = ['Cabin'], axis=1)
df_without_cabin.head()
```
